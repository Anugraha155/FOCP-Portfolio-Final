```python
# Greeting program with default for empty name
name = input("Enter your name: ").strip()
if name == "":
    print("Hello, Stranger!")
else:
    print(f"Hello, {name}!")
```

```python
# Basic password confirmation
p1 = input("Enter new password: ")
p2 = input("Confirm password: ")
if p1 == p2:
    print("Password Set")
else:
    print("Error: Passwords do not match")
```

```python
# Password length validation
p1 = input("Enter new password: ")
p2 = input("Confirm password: ")
if p1 == p2 and 8 <= len(p1) <= 12:
    print("Password Set")
else:
    print("Error: Invalid password")
```

```python
# Password check against common passwords
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']
p1 = input("Enter new password: ")
p2 = input("Confirm password: ")
if p1 == p2 and 8 <= len(p1) <= 12 and p1 not in BAD_PASSWORDS:
    print("Password Set")
else:
    print("Error: Invalid password")
```

```python
# Loop until valid password is chosen
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']
while True:
    p1 = input("Enter new password: ")
    p2 = input("Confirm password: ")
    if p1 == p2 and 8 <= len(p1) <= 12 and p1 not in BAD_PASSWORDS:
        print("Password Set")
        break
    else:
        print("Error: Try again")
        break
```

```python
# Seven times table
for i in range(13):
    print(f"{i} x 7 = {i*7}")
```

```python
# User-defined times table
n = int(input("Enter table number (0-12): "))
if 0 <= n <= 12:
    for i in range(13):
        print(f"{i} x {n} = {i*n}")
else:
    print("Invalid number")
```

```python
# Times table with reverse output for negative input
n = int(input("Enter table number: "))
if abs(n) <= 12:
    if n < 0:
        for i in range(12, -1, -1):
            print(f"{i} x {abs(n)} = {i*abs(n)}")
    else:
        for i in range(13):
            print(f"{i} x {n} = {i*n}")
else:
    print("Invalid number")
```
