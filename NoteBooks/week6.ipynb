```python
#colors method

colours = ["red", "green", "yellow", "blue", "red"]

print("Index of blue:", colours.index("blue"))
print("Count of red:", colours.count("red"))

copy_colours = colours.copy()
colours.append("black")

print("Original:", colours)
print("Copy:", copy_colours)
```

```python
# cubes comperhension

cubes = [x*x*x for x in range(2,21)]
print(cubes)
```

```python
# names sort ignoring case

names = ["Eric", "anna", "Sophie", "sam"]

names.sort(key=lambda n: n.upper())
print(names)
```

```python
# sorthing name

names = ["Eric", "anna", "Sophie", "sam"]

names.sort()
print(names)
```

```python
# Append next 3 squares
squares = [4, 9, 16, 25]
squares.append(36)
squares.append(49)
squares.append(64)

print(squares)
```

```python
# Extend with next 3 squares starting at 11Ã—11 = 121

squares = [4, 9, 16, 25, 36, 49, 64]
squares.extend([121, 144, 169])

print(squares)
```

```python
# Insert 2 at beginning

squares = [4, 9, 16, 25, 36, 49, 64]

squares.insert(0, 2)

print(squares)
```

```python
# square loop

import math

squares = [4, 9, 16, 25]

for s in squares:
    print(math.sqrt(s))
```

```python
# square pop

squares = [4, 9, 16, 25]

last = squares.pop()
print("Popped:", last)
print("List now:", squares)

first = squares.pop(0)
print("Popped first:", first)
print("List now:", squares)
```

```python
# removing squares

squares = [2, 4, 9, 16, 25, 36, 49, 64]

squares.remove(49)
print("After removing 49:", squares)


lst = [1,2,3,1,2]
lst.remove(2)  # removes first 2
print("Removing 2 from [1,2,3,1,2]:", lst)
```

```python
# reverse square

squares = [4, 9, 16, 25, 36]
squares.reverse()
print(squares)
```
