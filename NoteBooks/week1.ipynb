```python
# Simple arithmetic expression using variables
a = 45
b = 20
result = a + b
print(result)
```

```python
# Using arithmetic operators with variables
a = 10
b = 20
c = 15
print(a + b - c)

x = 10
y = 5
print(x * y)

m = 100
n = 33
print(m / n)
print(m // n)

p = 10
print(p ** 2)

q = 10
r = 3
print(q % r)
```

```python
# Operator precedence without parentheses using variables
a = 10
b = 5
c = 2
print(a + b * c)

x = 10
y = 5
z = 10
print(x - y * z + y)

m = 5
n = 10
print(m * n ** 2)
```

```python
# Operator precedence with parentheses using variables
a = 10
b = 5
c = 2
print((a + b) * c)

x = 10
y = 5
print(x - y * (x + y))

m = 5
n = 10
print((m * n) ** 2)
```

```python
# Nested parentheses precedence using variables
a = 12
b = 5
c = 2
d = 3

print(a + (b * c + d))
print(a + (b * (c + d)))
```
