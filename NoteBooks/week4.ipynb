```python
# task 4
"""Finds the maximum of two values."""
def findMax(a,b):
    if a > b:
        max = a
    else:
        max = b
    return max

print(findMax(3,7))
```

```python
# task 5
def multiply(a, b=None):
    if b is None:
        return a * a
    return a * b

print(multiply(3))
print(multiply(3,4))
```

```python
# task 8
def calcAve(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total/len(numbers)

print(calcAve(1,2,3))
```
