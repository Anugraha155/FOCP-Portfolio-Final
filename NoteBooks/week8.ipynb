```python
# Task 1: f-string rectangle area
w, h = 104.32, 15.654
area = w * h
print(f"The area of a rectangle with a width of {w} and a height of {h} is {area}")
```

```python
# Task 2: f-string with 3 decimal places
w, h = 104.32, 15.654
area = w * h
print(f"The area of a rectangle with a width of {w} and a height of {h} is {area:.3f}")
```

```python
# Task 3: alignment test
name = "Joe"
age = 75
print(f"{name:15} - {age:10}")
```

```python
# Task 4: right aligned 20 width, fill $, age 2dp
name = "Joe"
age = 75
print(f"{name:$>20} - {age:$>20.2f}")
```

```python
# Task 5: str.format() with keyword
import math
r = 52
area = math.pi * r * r
print("A circle with radius {} has an area of {area}".format(r, area=area))
```

```python
# Task 6: Convert f-string to str.format
name = "Donald"
age = 75
print("{name:@^15} - {age:#^10}".format(name=name, age=age))
```

```python
# Task 7: open, read all, close
f = open("info.txt")
content = f.read()
print(content)
f.close()
```

```python
# Task 8: readline() three times
f = open("info.txt")
print(f.readline())
print(f.readline())
print(f.readline())
f.close()
```

```python
# Task 9: for...in loop reading
f = open("info.txt")
for line in f:
    print(line)
f.close()
```

```python
# Task 10: append extra line
f = open("info.txt", "a")
f.write("\nthis is an extra line")
f.close()
```

```python
# Task 11: use with statement
with open("info.txt") as f:
    for line in f:
        print(line)
```
